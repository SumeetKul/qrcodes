```python
from manim import *  

config.media_width = "75%"
config.verbosity = "WARNING"
config.media_embed = True
```

```python
%%manim -v WARNING  --progress_bar None -ql --disable_caching Equation

class Equation(Scene):
    def construct(self):
        t0 = Text(r"[1, -2, 3, 5]", font_size = 66)
        t1 = Tex(r"$p(x) =$", r"$1x^3 -2x^2 +3x +5$", font_size = 66)
        t2 = Tex(r"$s(x) =$", r"$1x^5 - 2x^4 +3x^3 +5x^2$", r"$+ax + b$", font_size = 66)
        t3 = Text("redundant bits")
        g = VGroup(t1, t2, t3).arrange(DOWN, aligned_edge = LEFT, buff = 1)
        #g2 = VGroup(t2[2], t3).arrange(DOWN, aligned_edge = RIGHT, buff = 1)
        self.play(Write(t0), run_time = 4)
        self.remove(t0)
        self.wait(1)
        self.play(Write(t1), run_time = 4)
        self.wait(0.5)
        self.play(TransformFromCopy(t1[0], t2[0]))
        self.play(TransformFromCopy(t1[1], t2[1]), run_time=3)
        self.play(Write(t2[2]), run_time=2)
        t2.set_color_by_tex("a", GREEN)
        t2.set_color_by_tex("b", GREEN)
        self.wait(0.5)
        self.play(Write(t3), run_time=2, align=RIGHT)
        self.clear()
        t4 = Tex(r"$[1, -2, 3, $", r"$ 5, $", r"$ a, b]$", font_size = 66)
        t5 = Tex(r"$[1, -2, 3, $", r"$ 6, $", r"$ a, b]$", font_size = 66)
        t5.set_color_by_tex("6", RED)
        self.play(Write(t4), run_time = 4)
        self.remove(t4)
        self.add(t5)
        self.wait(1)
        
        self.wait()
```

```python
%%manim -v WARNING  --progress_bar None -ql --disable_caching Equation2

class Equation2(Scene):
    def construct(self):
        t1 = Tex(r"$p(x) =$", r"$1x^3 -2x^2 +3x +5$", font_size = 66)
        t2 = Tex(r"$p(x) \times x^2 = $", r"$1x^5 - 2x^4 +3x^3 +5x^2$", font_size = 66)
        t3 = Tex(r"$\frac{p(x) \times x^2}{g(x)}$", font_size = 66)
        t4 = Tex(r"$g(x) = (x-1)(x-2) = x^2 - 3x + 1$", font_size = 66)
        g = VGroup(t1, t2).arrange(DOWN, aligned_edge = LEFT, buff = 1)
        #g2 = VGroup(t2[2], t3).arrange(DOWN, aligned_edge = RIGHT, buff = 1)
        self.play(Write(t1), run_time = 4)
        self.wait(0.5)
        self.play(TransformFromCopy(t1[0], t2[0]))
        self.play(TransformFromCopy(t1[1], t2[1]), run_time=3)
        self.wait(1)
        self.clear()
        g = VGroup(t3, t4).arrange(DOWN, aligned_edge = LEFT, buff = 1)
        self.play(Write(t3))
        self.wait(0.5)
        self.play(Write(t4))
        self.wait(0.5)
        self.clear()
        t5 = Tex(r"$x^2 p(x) =$", r"$ q(x) * g(x)$", r"$ + r(x)$")
        t6 = Tex(r"$q(x) = x^3 + x^2 + 4 x + 15$")
        t7 = Tex(r"$r(x) = 37 x - 30$")
        g2 = VGroup(t5, t6, t7).arrange(DOWN, aligned_edge = LEFT, buff = 1)
        self.play(Write(t5), runtime=2)
        self.play(Write(t6), runtime=2)
        self.play(Write(t7), runtime=2)
        self.wait(2)
        self.clear()
        t8 = Tex(r"$s(x) = x^2 p(x) - r(x) = 1x^5 - 2x^4 +3x^3 +5x^2 - 37x + 30$")
        t9 = Tex(r"$a=-37, $", r"$b= 30$")
        g3 = VGroup(t8, t9).arrange(DOWN, aligned_edge = LEFT, buff = 1)
        self.play(Write(t8), runtime=2)
        self.play(Write(t9), runtime=2)
        self.wait()
```
